題目:<br>
讀取STOCK_DAY_0050_202009.csv串聯STOCK_DAY_0050_202010.csv，並且找出open小於close的資料


```python
import pandas as pd
```

```python
#讀取STOCK_DAY_0050_202009.csv、STOCK_DAY_0050_202010.csv
data09 = pd.read_csv('STOCK_DAY_0050_202009.csv')
data10 = pd.read_csv('STOCK_DAY_0050_202010.csv')

print(data09, '\n==========\n', data10)
```

```python
#串聯

#data09.keys() == data10.keys()
Mdata09_10 = pd.concat([data09, data10], axis=0)
print(Mdata09_10)
print('\n=========\n')
Mdata09_10_2 = pd.merge(data09, data10, on='date', how='outer')
print(Mdata09_10_2)
```

```python
#找出open小於close的資料

print(Mdata09_10[(Mdata09_10.open < Mdata09_10.close)])
print('\n=========\n')

Mdata09_10_2.loc[(Mdata09_10_2.open_x < Mdata09_10_2.close_x) | (Mdata09_10_2.open_y < Mdata09_10_2.close_y)]
```
